# Hypothesis Generation:


### Hypothesis 1: Does where the store is located affect how much money it makes? 
To test this, we compare the sales from stores in different locations, like Tier 1 and Tier 2, and see if they're               different or not.

### Hypothesis 2: Does the type of item being sold affect how much money the store makes? 
To test this, we compare the sales from different types of items, like perishables (like fruits and vegetables) and             non-perishables (like canned goods), and see if they're different or not.

### Hypothesis 3: Does how visible the items are on the store shelves affect how much money the store makes?
To test this, we compare the sales from items that are highly visible (with high visibility score) and less visible             (with low visibility score), and see if they're different or not.

### Hypothesis 4: Do discounts or prices of items affect how much money the store makes? 
To test this, we compare the sales from items with high discounts or prices (greater than 150 dollars) and low discounts               or prices (less than or equal to 150 dollars), and see if they're different or not.

To test these hypotheses, we use a statistical method called "t-test" for independent samples, which helps us compare two groups of data and see if they're different from each other. We calculate a number called "t-statistic" and a "p-value". If the p-value is less than 0.05 (which is a common threshold), then we say there's enough evidence to support the alternative hypothesis (meaning our idea is likely true). Otherwise, we say we don't have enough evidence to support our idea (meaning our idea is likely not true).

# Hypothesis Testing:

```python
# Hypothesis 1: Does location affect sales?
tier1_sales = train[train['Outlet_Location_Type'] == 'Tier 1']['Item_Outlet_Sales']
tier2_sales = train[train['Outlet_Location_Type'] == 'Tier 2']['Item_Outlet_Sales']

# equal_var =“True” :  t-test will be conducted by taking into consideration the equal population variances.
# equal_var =“False”:  t-test will be conducted by not taking into consideration the equal population variances.

t_stat, p_value = ttest_ind(tier1_sales, tier2_sales, equal_var=False)
alpha = 0.05 # significance level

print("Hypothesis 1: Does location affect sales?")
print("Null Hypothesis (H0): There is no significant difference in sales between Tier 1 and Tier 2 locations")
print("Alternate Hypothesis (H1): There is a significant difference in sales between Tier 1 and Tier 2 locations")
print("T-statistic: {:.2f}".format(t_stat)) # %.2f means to round up to two decimal places
print("P-value: {:.4f}".format(p_value))    # %.4f means to round up to two decimal places

if p_value < alpha:
    print("Conclusion: Reject Null Hypothesis")
    print("There is a significant difference in sales between Tier 1 and Tier 2 locations")
else:
    print("Conclusion: Fail to Reject Null Hypothesis")
    print("There is no significant difference in sales between Tier 1 and Tier 2 locations")
print()

# Hypothesis 2: Does item type affect sales?
perishable_sales = train[train['Item_Type'] == 'Perishable']['Item_Outlet_Sales']
non_perishable_sales = train[train['Item_Type'] == 'Non-Perishable']['Item_Outlet_Sales']

t_stat, p_value = ttest_ind(perishable_sales, non_perishable_sales, equal_var=False)

print("Hypothesis 2: Does item type affect sales?")
print("Null Hypothesis (H0): There is no significant difference in sales between perishable and non-perishable items")
print("Alternate Hypothesis (H1): There is a significant difference in sales between perishable and non-perishable items")
print("T-statistic: {:.2f}".format(t_stat))
print("P-value: {:.4f}".format(p_value))

if p_value < alpha:
    print("Conclusion: Reject Null Hypothesis")
    print("There is a significant difference in sales between perishable and non-perishable items")
else:
    print("Conclusion: Fail to Reject Null Hypothesis")
    print("There is no significant difference in sales between perishable and non-perishable items")
print()

# Hypothesis 3: Does item visibility affect sales?
high_visibility_sales = train[train['Item_Visibility'] > 0.1]['Item_Outlet_Sales']
low_visibility_sales = train[train['Item_Visibility'] <= 0.1]['Item_Outlet_Sales']

t_stat, p_value = ttest_ind(high_visibility_sales, low_visibility_sales, equal_var=False)

print("Hypothesis 3: Does item visibility affect sales?")
print("Null Hypothesis (H0): There is no significant difference in sales between high visibility and low visibility items")
print("Alternate Hypothesis (H1): There is a significant difference in sales between high visibility and low visibility items")
print("T-statistic: {:.2f}".format(t_stat))
print("P-value: {:.4f}".format(p_value))
print()
```

#### Handling Outliers:
An outlier is an extremely high or extremely low data point relative to the nearest data point and the rest of the neighboring co-existing values in a data graph or dataset.

```python
# Identify numerical column for outlier detection by seeing their datatypes
train.dtypes
```

```python
# List of column names used for outlier detection
num_col = ['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales'] 

# Create boxplots to visualize the distribution and identify outliers
plt.figure(figsize=(10, 6)) 
#It creates a new figure with a specific size (width=10, height=6) for the boxplots to be plotted.

# Loop through numerical columns and create boxplots
for col in num_col: # col is a variable that represents the current numerical column being processed in the loop
    plt.subplot(2, 2, num_col.index(col) + 1) # used to create subplots in a 2x2 grid arrangement for visualizing boxplots 
    sn.boxplot(x=train[col]) #  The x parameter specifies the data to be plotted on the x-axis, which is the values in the current numerical column.
    plt.xlabel(col)          # It sets the x-axis label of the boxplot to be the name of the current numerical column.

# Set the title of the entire subplot, which consists of multiple boxplots 
plt.suptitle('Boxplots for Numerical Columns (before handling outliers)')
plt.tight_layout() # automatically adjusts subplot params so that the subplots fit in to the figure area.
plt.show()         # display the boxplot
```

From the above box plots we can observe that Outliers are present in only Item_Visibility and Item_Outlet_Sales columns.

```python
def remove_outlier(col):
    col_sorted = sorted(col)
    Q1, Q3 = col.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range

lower_itemsales, higher_itemsales = remove_outlier(train['Item_Outlet_Sales'])

# Replace values greater than upper bound with upper bound value
train['Item_Outlet_Sales'] = train['Item_Outlet_Sales'].where(train['Item_Outlet_Sales'] <= higher_itemsales, higher_itemsales)

# Replace values lower than lower bound with lower bound value
train['Item_Outlet_Sales'] = train['Item_Outlet_Sales'].where(train['Item_Outlet_Sales'] >= lower_itemsales, lower_itemsales)

sn.boxplot(y=train['Item_Outlet_Sales'])
plt.title('Item Outlet Sales Distribution after removing outliers')
```
